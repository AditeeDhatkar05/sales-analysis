```python
# As ExternOrderNo and Order ID are same so I renamed it with Order ID column.
OrderReport_SKUMaster=Order_Report.merge(right=SKU_Masters,on="SKU")
OrderReport_SKUMaster=OrderReport_SKUMaster.rename(columns={"ExternOrderNo":"Order ID"})
OrderReport_SKUMaster.head()
```

```python
# As per requirment, we want weights are in the form of KG so I converted it into KG form.
OrderReport_SKUMaster["Total Weight(g)"]=OrderReport_SKUMaster["Order Qty"]*OrderReport_SKUMaster["Weight (g)"]
OrderReport_SKUMaster["Total Weight(KG) as per X"]=OrderReport_SKUMaster["Total Weight(g)"]/1000
OrderReport_SKUMaster.head()
```

```python
# Just sliced the columns which I need.
data1=OrderReport_SKUMaster[["Order ID","Order Qty","Total Weight(KG) as per X"]]
data1.head()
```

```python
# Merged two datasets on common column order ID
df_merged = pd.merge(data1, Courier_Comp_invoice[["Order ID","AWB Code"]], on='Order ID', how='inner')
df_merged.head()
```

```python
#Aggrigate Numerical Columns
df_summed = df_merged.groupby('Order ID').agg({'Order Qty':'sum','Total Weight(KG) as per X':'sum','AWB Code': 'first',}).reset_index()
df_summed.head()
```

```python
final_df=df_summed[["Order ID","AWB Code","Order Qty","Total Weight(KG) as per X"]]
final_df.head()
```

```python
pinzone_invoice=pd.merge(X_pincode_zone,Courier_Comp_invoice[["Customer Pincode","Zone","Order ID"]],
                         on="Customer Pincode",how="left")
pinzone_invoice.head()
```

```python
data2=pd.merge(final_df,pinzone_invoice,on="Order ID",how="inner")
data2.head()
```

```python
data2=data2.drop_duplicates()
data2.head()
```

```python
# Adding Zone columns of X company and Courier company.
data2=data2.rename({"Zone_x":"Zone"},axis=1)
data2.head()
```

```python
# Added new column in dataframe which is Weight slab of X.
data3= pd.merge(data2, Courier_Comp_rate[['Zone', 'Weight Slabs']], on='Zone', how='left')
data3['Weight Slab As Per X'] = data3['Weight Slabs']
data3 = data3.drop(columns=['Weight Slabs'])
data3
```

```python
data4=data3.rename({"Zone":"Delivery Zone As per X"},axis=1)
data4.head()
```

```python
data4=data4.rename({"Zone_y":"Zone"},axis=1)
data4.head()
```

```python
# Added Weight slab column which is used by Courier company.
data5= pd.merge(data4, Courier_Comp_rate[['Zone', 'Weight Slabs']], on='Zone', how='left')
data5['Weight Slab charged by Courier Company'] = data5['Weight Slabs']
data5 = data5.drop(columns=['Weight Slabs'])
data5
```

```python
# Renamed it as per our requirment
data5=data5.rename({"Zone":"Delivery Zone Charged By Courier Company"},axis=1)
data5.head()
```

```python
# Merged datasets for total weights by courier company
data6=pd.merge(data5,Courier_Comp_invoice[["Charged Weight","Order ID"]],on="Order ID",how="inner")
data6.head()
```

```python
data6=data6.rename({"Charged Weight":"Total Weight As Per Courier Company"},axis=1)
data6.head()
```

```python
data7=pd.merge(data6,Courier_Comp_invoice[["Type of Shipment","Order ID"]],on="Order ID",how="inner")
data7.head()
```

```python
data8 = pd.merge(data7, Courier_Comp_rate, left_on='Delivery Zone As per X', right_on='Zone', how='left')
data8.head()
```

```python
data8["Type of Shipment"].value_counts()
```

```python
# Calculated as per logic, Billing Amount based on Type of Shipment
def calculate_billing_amount(row):
    
    if 'Forward charges' in row['Type of Shipment']:
        fixed_charge = row['Forward Fixed Charge']
        additional_charge = row['Forward Additional Weight Slab Charge']
        return fixed_charge+additional_charge
        
    else:
        fixed_charge = row['Forward Fixed Charge']
        rto_charges=row['RTO Fixed Charge']
        additional_charge = row['RTO Additional Weight Slab Charge']
        return fixed_charge+rto_charges+additional_charge
    

data8['Billing Amount of X'] = data8.apply(calculate_billing_amount, axis=1)

data8.head()
```

```python
data9=pd.merge(data8,Courier_Comp_invoice[["Billing Amount (Rs.)","Order ID"]],on="Order ID", how="inner")
data9.head()
```

```python
# Calculated difference between X company charges and Courier company charges.
data9["Diff_Billing"]=data9["Billing Amount (Rs.)"]-data9["Billing Amount of X"]
data9 = data9.round(2)
data9.head()
```

```python
data9.columns
```

```python
data9=data9.rename({"Billing Amount of X":"Expected Charges as per X(Rs.)"},axis=1)
data9=data9.rename({"Billing Amount (Rs.)":"Charges Billed by Courier Company(Rs.)"},axis=1)
```

```python
# Extracting Columns which we required
Output_one_final_df=data9[['Order ID', 'AWB Code','Total Weight(KG) as per X','Weight Slab As Per X',
                          'Total Weight As Per Courier Company','Weight Slab charged by Courier Company',
                           'Delivery Zone As per X','Delivery Zone Charged By Courier Company','Expected Charges as per X(Rs.)',
                          'Charges Billed by Courier Company(Rs.)','Diff_Billing']]
```
